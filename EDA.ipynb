```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
entree_df = pd.read_csv('mc_entree_with_price.csv')
entree_df = entree_df.drop(columns = 'Time').rename(columns ={ 'Unnamed: 0': "Item_Name"})
entree_df
```

# Viewing Correlations to Calories

```python
sns.heatmap(entree_df.corr())
```

From this heatmap, I can see that none of the items have any Vitamin D, and that the strongest connections to calories seems to be Total Fat, Protein, and surprisingly Potassium.

```python
sns.scatterplot(x=entree_df['Calories'], y=entree_df['Protein'])
plt.title('Correlation between Protein and Calories');
```

```python
sns.scatterplot(x=entree_df['Calories'], y=entree_df['Potassium'])
plt.title('Correlation between Potassium and Calories');
```

```python
sns.scatterplot(x=entree_df['Calories'], y=entree_df['Total_Fat'])
plt.title('Correlation between Total Fat and Calories');
```

```python
sns.scatterplot(x=entree_df['Calories'], y=entree_df['Calcium'])
plt.title('Calories to Calcium Comparison');
```

Calcium does not seem to be connected to calories, having the lowest correlation. This is probably because it is primarily coming from dairy products like cheese

```python
calsproteinratio = []
for item in range(len(entree_df['Calories'])):
    cals = entree_df["Calories"][item]
    protein = entree_df["Protein"][item]
    ratio = protein/cals
    calsproteinratio.append(ratio)
    
entree_df['Cals_Protein_Ratio'] = calsproteinratio
entree_df
```

```python
bar = sns.barplot(x='Item_Name', y='Cals_Protein_Ratio', data=entree_df);
plt.xticks(rotation=90);
```

## From this barplot, we can see that the double quarter pounder with cheese has the most protein per calorie, while the mcchicken has the least

```python
pricepercal = []
for item in range(len(entree_df['Calories'])):
    cals = entree_df["Calories"][item]
    price = entree_df["Price"][item]
    ratio = price/cals
    pricepercal.append(ratio)
    
entree_df['Price_Per_Cal'] = pricepercal
entree_df
```

```python
bar = sns.barplot(x='Item_Name', y='Price_Per_Cal', data=entree_df);
plt.xticks(rotation=90);
```

## However, here we see that the mcchicken is the best price for the calories, while the filet-o-fish is the most expensive per calorie. (The McChicken is .6 cents per calorie, the Filet-o-Fish is 1.5 cents per calorie)

```python
priceperprot = []
for item in range(len(entree_df['Protein'])):
    protein = entree_df["Protein"][item]
    price = entree_df["Price"][item]
    ratio = price/protein
    priceperprot.append(ratio)
    
entree_df['Price_Per_Protein'] = priceperprot
entree_df
```

```python
bar = sns.barplot(x='Item_Name', y='Price_Per_Protein', data=entree_df);
plt.xticks(rotation=90);
```

## When looking at the price per gram of protein, once again the filet-o-fish shows itself to be the most expensive. The double cheesburger, however, is the cheapest
